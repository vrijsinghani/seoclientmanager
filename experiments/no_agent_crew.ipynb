```python
from crewai import Agent, Task, Crew, Process
from crewai_tools import tool

# Define a custom tool that performs a procedural operation
@tool
def simple_greeting(name: str) -> str:
    """
    This tool returns a greeting message for the given name.
    
    Args:
        name (str): The name of the person to greet.
    
    Returns:
        str: A greeting message.
    """
    return f"Hello, {name}!"

# Define a simple "dummy" agent to execute the task
dummy_agent = Agent(
    role='Simple Executor',
    goal='Execute simple tasks without complex reasoning.',
    backstory='a simple agent that does not use an llm',
    tools=[simple_greeting],  # Assign the procedural tool to the agent
    memory=False,  # No memory needed
    reasoning=False  # No reasoning involved
)

# Define a task that uses the agent to run the tool
greeting_task = Task(
    description='Greet a user by their name.',
    expected_output='A greeting message',
    tools=[simple_greeting],  # Using the tool directly
    agent=dummy_agent  # Assigning the dummy agent
)

# Create a crew that executes this task
crew = Crew(
    agents=[dummy_agent],  # Use the dummy agent
    tasks=[greeting_task],  # Only the procedural task
    process=Process.sequential  # Sequential execution
)

# Kick off the crew with inputs for the task
result = crew.kickoff(inputs={'name': 'Alice'})

# Output the result
print(result)
```
